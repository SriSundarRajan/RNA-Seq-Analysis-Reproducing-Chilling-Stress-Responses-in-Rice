# Bash & Bioinformatics Cheat Sheet
*Quick reference for terminal navigation, the SRA Toolkit, and FastQC.*

---

## 1. Core Shell & File System Fundamentals

- `/` — Root directory of the system
- **Absolute path** — full path starting from root (e.g. `/home/user/data`)
- **Relative path** — path relative to your current location (e.g. `../data`)

| Command | Description / Usage |
|---|---|
| `pwd` | Print the present working directory (absolute path). |
| `cd [path]` | Change directory to an absolute or relative path. |
| `cd ~` | Return directly to your home directory. |
| `cd ..` | Move up one level to the parent directory. |
| `ls` | List files and folders in the current directory. |
| `ls -l` | Long format (sizes, permissions, dates). |
| `ls -a` | Show all contents, including hidden files (e.g. `.bashrc`). |

---

## 2. File & Directory Management

| Command | Description / Usage |
|---|---|
| `mkdir [dir]` | Create a new directory. |
| `touch [file]` | Create an empty file, or update an existing file's timestamp. |
| `vim [file]` | Open/edit a file in the Vim text editor. |
| `nano [file]` | Open/edit a file in the simple, beginner-friendly editor. |
| `cp file1 file2` | Copy `file1` to a new file named `file2`. |
| `mv old new` | Move or rename a file or directory. |
| `rm [file]` | Permanently remove/delete a file. |
| `rmdir [dir]` | Remove an **empty** directory. |

---

## 3. Input/Output, Combining & Viewing Files

| Operator / Command | Description & Purpose |
|---|---|
| `*` | Wildcard: matches any sequence of characters (e.g. `*.fasta`). |
| `>` | Redirect output: overwrites the target file, or creates it. |
| `>>` | Append output: adds to the end of an existing file. |
| `cat [file]` | Print full file content to standard output. |
| `less [file]` | View large files page-by-page (arrows to scroll, `q` to quit). |

---

## 4. Text Processing, Extraction & Searching

| Command Syntax | Data Parsing Action |
|---|---|
| `grep "pattern" file` | Print lines in `file` that match the pattern. |
| `grep -c "pattern" file` | Count the number of matching lines. |
| `grep -v "pattern" file` | Invert match: print lines that do **not** match. |
| `head -n N file` | Show the first `N` lines (default 10 if `-n` omitted). |
| `tail -n N file` | Show the last `N` lines (default 10). |
| `cut -f1,3 file` | Extract columns 1 and 3 from tab-delimited files. |
| `cut -d "," -f1,3 file` | Extract columns using a custom delimiter (e.g. CSV). |

---

## 5. Terminal Shortcuts & Session Control

| Shortcut / Tool | Effect |
|---|---|
| `↑` / `↓` arrows | Scroll through command-line history. |
| `TAB` | Auto-complete file, directory, or tool names. |
| `Ctrl + C` | Kill/terminate the current foreground process. |
| `Ctrl + L` | Clear the terminal screen. |
| `man [command]` | Display the offline manual for a command. |
| `exit` | End the current shell session. |

## 6. Next-Gen Sequencing (NGS) Tools

### SRA Toolkit — `fastq-dump`

| Option | Action / Result |
|---|---|
| `fastq-dump -h` | Print help documentation and parameters. |
| `fastq-dump [SRR_ID]` | Download and unpack a complete run into raw FASTQ. |
| `fastq-dump -X [N] [SRR_ID]` | Output only the first `N` spots (great for quick tests). |
| `fastq-dump --skip-technical` | Dump biological reads only, omitting technical reads (barcodes/primers/linkers). |
| `fastq-dump -Z [SRR_ID]` | Stream output to stdout (the screen). |
| `fastq-dump -F` | Retain original read names/defline format. |
| `fastq-dump --split-files` | Split paired-end reads into separate `_1.fastq` and `_2.fastq` files. |
| `fastq-dump --gzip` | Compress output with **gzip** (`.fastq.gz`). |

> **Note:** `fastq-dump` is the legacy tool. For most downloads, `fasterq-dump` is faster (multi-threaded), though it lacks a few flags such as `-Z` and inline `--gzip`.

### FastQC & Regex

| Command / Pattern | Property & Interpretation |
|---|---|
| `fastqc [file]` | Run quality-control stats; generates an interactive HTML report. |
| `-t [threads]` | Use multiple threads to process files in parallel. |
| `-o [dir]` | Set the output directory for HTML results. |
| `--nogroup` | Disable base-position binning for reads longer than 50 bp. |
| `"^>"` | Regex: start of a header line in a FASTA file. |
| `"^@"` | Regex: start of an identifier line in a FASTQ file. |

# Git Basics

### Basic commands:

```bash
git clone [Git-Project-URL]
```
- Downloads Project to working directory, including git-exclusive files

```bash
git init 
```
- "Initializes" current working directory as a git project (creates ".git"-folder)
- No existing project needed, works completely locally

```bash
git add .
```
- Adds all files to the file-list
- The "." represents "current directory", paths and single files work as well

```bash
git commit -m "[MESSAGE]"
```
- "Commit" to the changes you made and the files you added
- Message is mandatory and should always contain proper information
- Saves current state of changes on the local machine, NOT online

```bash
git push origin 
```
- Uploads the Repository to the Server
- Also compares if there are potential conflicts or outdates versions
- Requires either stored login details or a manual login afterwards
```bash
git status
```
- Show all changes since the last commit
- Also shows you the currently active "branch"
```bash
git switch -b [BRANCH-NAME]
```
- Create new branch with specified name or change to existing one
```bash
git checkout [BRANCH/OLD-VERSION]
```
- Changes current version to another branch (like switch)
- Alternatively, change to old version by providing version-hash
```bash
git restore "[FILE/PATH-NAME]"
```
- Undo changes, as added via "git add"
```bash
git merge [BRANCH]
```
- "Merge" branch back to the main branch
- Basically, commits all of the commits in "branch" to main
- Should merge-errors arise: use "git mergetool"

## Use-Cases & Examples
### Download, copy and upload
Like in this course: 
- Download the data from one repository
```bash
git clone [REPO-NAME]
```
- Move all the files to a new folder
```bash
cd ..
mkdir [NEWREPO]
cp -r [OLDREPO]/[FOLDER] [NEWREPO] 
# You can also copy everything with "cp [OLDREPO]/* [NEWREPO]/" and then delete the ".git" folder "rm -r [NEWREPO]/.git"
cd [NEWREPO]
```
- Define this as a new git project and add existing files
- (You would start here if you start a project from scratch)
```bash
git init
git add .
git commit -m "Initial commit from existing project"
```
- Upload this to the remote(/web) location
```bash
git remote add origin [YOUR-URL]
git branch -M main
git push -u origin main
```
- You will be prompted to log in/authenticate

### Collaborate on an existing project (without branches)
First, make sure all collaborators are members of the git group/project
- Then, Download the project
```bash
git clone [REPO-NAME]
```
- Make changes to existing files, delete stuff or add new ones (no code samples for this)
- Add and commit changes (you can do multiple commits for separate changes)
```bash
git add .
git commit -m "Changed X, added Y, fixed Z"
```
- Push to the remote server
```bash
git push -u origin main
```
- You will be prompted to log in/authenticate
