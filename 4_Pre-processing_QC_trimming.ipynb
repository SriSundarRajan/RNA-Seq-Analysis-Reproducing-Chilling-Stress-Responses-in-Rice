### Adapter trimming with fastp
**`fastp`** is a fast, all-in-one, ultra-efficient tool for preprocessing FASTQ files in high-throughput sequencing data. It is widely used for **quality control (QC)** and **adapter trimming** in NGS pipelines.

### Key Features:
- **Quality filtering:** Removes low-quality reads or bases.
- **Adapter trimming:** Automatically detects and removes adapter sequences.
- **Base correction:** Improves data quality by correcting mismatched bases.
- **Poly-G tail trimming:** Especially useful for Illumina NovaSeq data.
- **Output in multiple formats:** FASTQ, JSON, and HTML reports for easy visualization.

### Why Use `fastp`?
- **Speed:** Parallel processing makes it extremely fast.
- **Simplicity:** Single command to perform multiple preprocessing steps.
- **Comprehensive QC:** Generates detailed reports in both JSON and interactive HTML.

### Other adapter trimming software and When to Use Them:

1. **Trim Galore** – Best for **adapter trimming** and **quality filtering**, especially if you're working with **Illumina** data and require specific trimming strategies like removing methylation bias in **bisulfite sequencing**.

2. **Trimmomatic** – Great for **customized trimming workflows**; offers more granular control over operations like sliding window trimming and palindrome clip detection. Ideal for **large datasets** where flexibility is needed.

3. **Cutadapt** – Focused on **adapter removal** with **custom adapter sequences**. It's highly flexible and supports **RNA-seq and amplicon sequencing**, where adapter sequences can vary.


```python
%%bash
mkdir -p /mnt/volume/data/group8/studies/rice/run/trimmed_reads
for fastq in /mnt/volume/data/group8/studies/rice/*1.fastq; do
    basename=$(basename $fastq _1.fastq)
    in_name=$(dirname $fastq)
    out_name=/mnt/volume/data/group8/studies/rice/run/trimmed_reads
    R1=$in_name/${basename}_1.fastq
    R2=$in_name/${basename}_2.fastq
    O1=$out_name/${basename}_1.fastq
    O2=$out_name/${basename}_2.fastq
    
    echo "Processing $basename with fastp..." 
    fastp -i $R1 -I $R2 -o $O1 -O $O2 && \
    echo "Finished processing $basename." && \
    rm $R1 $R2
done
```
