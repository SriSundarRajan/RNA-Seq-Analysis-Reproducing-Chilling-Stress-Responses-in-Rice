
### Step 1: Install SSH Client
First, ensure that everyone has an SSH client installed on their system.

- **Mac**: macOS comes with SSH pre-installed.
- **Windows**: We recommend installing Windows Subsystem for Linux (WSL) or using PuTTY if you prefer a GUI. Windows 10 and later also include OpenSSH client which can be enabled through the "Optional Features" settings.
- **Linux**: Most Linux distributions come with SSH pre-installed. If not, it can be installed via the package manager (e.g., `sudo apt install openssh-client` on Ubuntu).

### Step 2: Setup SSH Agent for Users

#### For PowerShell Users:
1. **Start the SSH Agent**:
   - Open PowerShell as an administrator.
   - Execute the following command to start the SSH agent service if it's not already running:
     ```powershell
     Start-Service ssh-agent
     ```
   - If the service is not installed, you may need to enable it via Windows features or install OpenSSH Client from the "Optional Features" in Settings.

2. **Add the Private Key to the Agent**:
   - Download the private key from here: [login_student](../../login_student)
   - Use the full Windows path to add the SSH key:
     ```powershell
     ssh-add C:\path\to\login_student
     ```
   - Ensure that the path to the `login_student` file is correct and replace `C:\path\to\ login_student` with the actual path where your key is stored.

#### For WSL and Linux Users:
1. **Start the SSH Agent**:
   - Open your WSL terminal.
   - It's possible that the `ssh-agent` isn't running in the background of your current shell session, in any case, re-running this command wouldn't hurt.:
     ```bash
     eval "$(ssh-agent -s)"
     ```
   - This command starts `ssh-agent` in the background and ensures the environment variables set by `ssh-agent` are correctly added to your current shell session, allowing `ssh-add` to communicate with it.

2. **Add Your Private Key to the SSH Agent**:
   - Ensure your private key file (`login_student`) downloaded from here [login_student](https://gitlab.lrz.de/computational_plant_biology/courses/FunCourse/-/blob/main/login_student), is accessible within the WSL file system. If it's stored on your Windows system, you might access it via `/mnt/c/path/to/login_student`.
   - Add the SSH key to the agent:
     ```bash
     ssh-add /path/to/login_student
     ```
   - Replace `/path/to/login_student` with the actual path to your key.

### Additional Notes for WSL Users:
- **Persistent Agent Across Sessions**: To keep the `ssh-agent` running across different terminal sessions, you might want to add the `eval "$(ssh-agent -s)"` command to your shell's startup file like `~/.bashrc` or `~/.zshrc`.
- **File Permissions**: Ensure that the permissions for `login_student` are appropriately set to allow only the user read access, typically with `chmod 600 /path/to/login_student`.

These detailed steps should help Windows users, both in PowerShell and WSL environments, to set up and use the SSH agent effectively for their SSH connections.

### Troubleshooting Common Issues
- **Permission Denied**: Ensure the private key file’s permissions are correctly set (`chmod 600 /path/to/login_student` on Mac/Linux).
- **Agent Has No Identities**: Make sure the key is correctly added to the SSH agent with `ssh-add -l` to list loaded keys. WSL users may have to shutdown fully their wsl or windows machine and restart it all.
- **Connectivity Issues**: Verify network settings, ensure the IP addresses and hostnames in the SSH command are correct and reachable.

### Step 4: Setting up VS Code Remote Development

###  Installing Conda Environment from a YAML Package
Once you have set up SSH and VS Code, the next step is to install a Conda environment from a `.yml` package. This allows you to recreate the environment with the exact dependencies specified in the YAML file. Here's how you can proceed:
1. **Install Miniconda (If Conda is Not Installed)**
 If Conda is not already installed on the remote server, you can install **Miniconda**, a lightweight version of Conda that includes just the Conda package manager and Python.
   - **Download Miniconda**:
     On the remote server, download Miniconda for your system (e.g., Linux) by running:
     ```bash
     wget https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh
     ```
   - **Install Miniconda**:
     Run the following command to start the Miniconda installation:
     ```bash
     bash Miniconda3-latest-Linux-x86_64.sh
     ```
     Follow the prompts during the installation process. Afterward, you might need to close and reopen your terminal or run:
     ```bash
     source ~/.bashrc
     ```
   - **Verify Miniconda Installation**:
     Ensure Miniconda is correctly installed by running:
     ```bash
     conda --version
     ```

2. **Create the Environment from the YAML File**

   Once Miniconda or Conda is set up and you have cloned this Git repository to your remote server, you can create a Conda environment from the `RNAseq.yml` package.
   Use the following command to create the Conda environment:
   ```bash
   conda env create -f /path/to/RNAseq.yml
   ```
   Replace `/path/to/RNAseq.yml` with the actual path to your YAML file on the remote server.

3. **Activate the New Conda Environment**

   After the environment is created, activate it by running:
   ```bash
   conda activate RNAseq
   ```

4. **Install Additional Dependencies (If Needed)**

   If there are any additional dependencies or steps needed that aren't included in the `environment.yml` file, you can manually search for packages on the [Anaconda page](https://anaconda.org/) and install them using Conda:
   ```bash
   conda install <package_name>
   ```
   Replace `<package_name>` with the name of the package you wish to install.

5. **Verify Installation**

   After setting up the environment, verify that everything is correctly installed by running:
   ```bash
   conda list
   ```
   This will display the list of installed packages in your environment.
