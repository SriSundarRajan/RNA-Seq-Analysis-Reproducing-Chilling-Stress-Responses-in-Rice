# Aligning to Transcriptome with Kallisto.
---

## Kallisto Alignment
**Kallisto** uses a different approach: **pseudoalignment**, which is much faster than traditional aligners like STAR. Instead of aligning reads to the genome, it matches reads to a set of **transcripts** using **k-mer counting**.  
- **Output:** Direct transcript abundance estimates (TPM, FPKM, counts).  
- **Use case:** When speed and transcript-level quantification are priorities, especially for **well-annotated transcriptomes**.

## STAR Alignment
**STAR (Spliced Transcripts Alignment to a Reference)** is a high-performance aligner designed specifically for **RNA-Seq data**. It maps sequencing reads to a reference genome, allowing for **splice junction detection** and handling of large, complex genomes.  
- **Output:** Genome-aligned BAM/SAM files.  
- **Use case:** When you want to identify **splice variants, novel transcripts**, or quantify gene expression at the **exon and intron** level.

---

### RNA-Seq Quantification
**Gene and transcript quantification** from RNA-Seq data can be done in two primary ways:  
1. **STAR + FeatureCounts or HTSeq**: Align reads with STAR, then count mapped reads to genes for accurate quantification.  
2. **Kallisto (or Salmon)**: Direct quantification without full alignment, providing **faster and transcript-level estimates**.

**When to use which:**  
- **STAR:** For **detailed genome-level analysis** (e.g., novel isoforms, splice variants).  
- **Kallisto:** For **speed and transcript-level quantification** in large datasets or **well-annotated species**.

## Summarizing findings to gene level count tables.
---

Kallisto (or Salmon) quantifies **transcript-level abundance**, meaning it provides counts for individual transcripts (isoforms) rather than genes. However, for many downstream analyses (e.g., differential gene expression using `DESeq2` or `edgeR`), you typically want gene-level counts, not transcript-level counts.

The `tx2gene` file provides a **mapping between transcripts and genes**, which allows `tximport` to:
1. **Aggregate transcript-level counts to gene-level counts**.
2. **Summarize abundances** (like TPM) at the gene level.
3. **Improve estimates of gene expression** by correcting for biases related to transcript length and isoform expression.

Without `tx2gene`, `tximport` would have no way of knowing which transcripts belong to which genes.

---

### **How `tximport` Works with `tx2gene`**

1. **Input**:
   - A set of `abundance.tsv` files from Kallisto, each containing transcript-level data for every sample.
   - The `tx2gene` file tells `tximport` how transcripts (e.g., `TraesCS1A02G000100.1`) are grouped into genes (e.g., `TraesCS1A02G000100`).
   
2. **Summarization to Gene Level**:
   - `tximport` reads the transcript counts, abundances (TPM), and lengths for each sample.
   - For each gene, it sums up the counts and averages the lengths of its associated transcripts (weighted by abundance).
   - This process yields gene-level **counts**, **TPM**, and **average transcript length**, which are used to normalize counts for downstream analysis.

---

### **Why Gene-Level Counts are Important**
1. **Biological Relevance**:
   - Genes, not individual transcripts, are the primary unit of interest in most expression studies.
   - Gene-level analysis is easier to interpret, especially when comparing biological conditions (e.g., disease vs. control).

2. **Robustness**:
   - Summarizing to the gene level reduces the noise caused by alternative splicing and incomplete transcript annotations.
   - It mitigates problems caused by poorly annotated or novel transcripts that might have low read support.

3. **Compatibility with Downstream Tools**:
   - Most differential expression tools (`DESeq2`, `edgeR`) require gene-level counts as input.
   - Gene-level counts simplify normalization and statistical modeling, as genes typically have more stable read counts than transcripts.

---

### **Example Workflow Recap**

1. **Raw Data**: Fastq files for multiple samples.
2. **Kallisto**: Quantifies transcript-level expression for each sample and outputs `abundance.tsv`.
3. **`tximport` with `tx2gene`**:
   - Reads the `abundance.tsv` files.
   - Uses `tx2gene` to map transcripts to their corresponding genes.
   - Summarizes counts to the gene level.
4. **Gene-Level Count Matrix**: Ready for downstream analysis with `DESeq2` or `edgeR`.

---

### **Visualization of the Process**


---

### Why This Helps
- **Accurate gene-level quantification** improves your statistical power.
- **Consistent data format** for use with standard analysis tools.
- **Biological interpretation** becomes easier at the gene level.

````r
# Load the libraries quietly
library(rtracklayer, quiet = TRUE)
library(dplyr, quiet = TRUE)

# Define the path to your GFF file
gff_file <- "/mnt/volume/data/group8/studies/rice/references/osa1_r7.all_models.gff3"

# Read the GFF file
gff <- import(gff_file)

# Check column names in the GFF file
column_names <- colnames(as.data.frame(gff))
print(column_names)
````

````r
# Import GFF file
gff <- import(gff_file)
head(gff[gff$type == "mRNA", ])
````

````r
# Convert CharacterList to character vector
gff_tx <- gff %>%
  as.data.frame() %>%
  filter(type == "mRNA", !is.na(ID), !is.na(Parent))
  #filter(type == "mRNA" & !is.na(transcript_id) & !is.na(Parent))


# Convert Parent (CharacterList) to a clean character vector
gff_tx$Parent <- sapply(gff_tx$Parent, function(x) as.character(x[1]))

# Remove "gene:" prefix from Parent if present
#gff_tx$Parent <- sub("gene:", "", gff_tx$Parent)

# Build tx2gene data frame
tx2gene <- gff_tx %>%
  dplyr::select(ID, Parent) %>%
  rename(gene_id = Parent) %>%
  distinct()

# Save tx2gene to a CSV file
write.csv(tx2gene, "/mnt/volume/data/group8/studies/rice/references/tx2gene.csv", row.names = FALSE)

# Print a preview of the tx2gene file
head(tx2gene)
````

````r
library(tximport)

# Define paths to trimmed reads and extract sample names
fastq_files <- list.files(path = "/mnt/volume/data/group8/studies/rice/run/trimmed_reads", pattern = "*_1.fastq$", full.names = TRUE)
samples <- basename(fastq_files) %>% sub("_1\\.fastq$", "", .)

# Create paths to Kallisto abundance files
files <- file.path("/mnt/volume/data/group8/studies/rice/run/kallisto_output", samples, "abundance.tsv")
names(files) <- samples

# Check that all files exist
missing_files <- files[!file.exists(files)]
if (length(missing_files) > 0) {
  cat("Warning: The following abundance.tsv files are missing:\n")
  print(missing_files)
} else {
  cat("All abundance.tsv files found.\n")
}

# Load the tx2gene mapping
tx2gene <- read.csv("/mnt/volume/data/group8/studies/rice/references/tx2gene.csv")

# Run tximport to summarize counts to gene level
txi <- tximport(files, type = "kallisto", tx2gene = tx2gene)

# Check the structure of the imported object
str(txi)

# Save gene-level counts to a CSV file
write.csv(txi$counts, file = "/mnt/volume/data/group8/studies/rice/run/kallisto_output/gene_counts.csv", row.names = TRUE)
write.csv(txi$abundance, file = "/mnt/volume/data/group8/studies/rice/run/kallisto_output/gene_counts_abundance.csv", row.names = TRUE)
cat("Gene-level counts saved to /mnt/volume/data/group8/studies/rice/run/kallisto_output/gene_counts.csv\n")
````
