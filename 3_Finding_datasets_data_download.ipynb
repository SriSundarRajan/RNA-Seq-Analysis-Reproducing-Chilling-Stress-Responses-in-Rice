# Exploring the SRA (Sequence Read Archive) Database
The Sequence Read Archive (SRA) stores raw sequencing data from the scientific community and provides a robust interface for searching these data. Follow this link to log onto the  [Advanced Search - SRA - NCBI](https://www.ncbi.nlm.nih.gov/sra/advanced).


## Key Components of the SRA Search Builder:
**Organism:** Reference genome file for rice 

**Platform:** Chosen as Illumina, known for its widespread use and data standardization, which makes it ideal for our studies.
Why Illumina?: It offers standardized outputs that are compatible with most bioinformatics tools, making it a practical choice for learning and initial research explorations.

**Strategy Field:** Lists various sequencing strategies (e.g., RNA-Seq, ChIP-Seq). This helps in refining searches according to the experimental approach or study focus. But as the number of sequencing approaches increase, the data standardization element is weaker here. Please know that this filter may not include projects you are actually looking.

## Install SRA Toolkit
From here, we communicate between SRA and our compute node directly and programmatically.


Download [SRA toolkits](https://github.com/ncbi/sra-tools/wiki/01.-Downloading-SRA-Toolkit#sra-toolkit), and unpack it. This link is also a tutorial to help download SRA-files. To call the Sra toolkit in your current shell, you will need to edt the <PATH> variable. Below is the example for Ubuntu systems.

```
wget -P ~ https://ftp-trace.ncbi.nlm.nih.gov/sra/sdk/3.2.0/sratoolkit.3.2.0-ubuntu64.tar.gz
tar -xvf ~/sratoolkit.3.2.0-ubuntu64.tar.gz -C ~
export PATH=$PATH:~/sratoolkit.3.2.0-ubuntu64/bin
echo 'export PATH=$PATH:~/sratoolkit.3.2.0-ubuntu64/bin' >> ~/.bashrc
```

Then [configure the sra toolkit](https://github.com/ncbi/sra-tools/wiki/03.-Quick-Toolkit-Configuration) with the instructions [here](https://github.com/ncbi/sra-tools/wiki/05.-Toolkit-Configuration).
now you are ready to start downloading files.
